### Q5/6 Does this coffee shop make more money from selling drinks or from selling food?

```python
f_counter = 0
f_sum = 0
d_counter = 0
d_sum = 0

for i, row in df.iterrows():
    if row["Item"] in three_groups_dict["food"]:
        f_counter += 1
        f_sum += row["Item_Price"]
    elif row["Item"] in three_groups_dict["drinks"]:
        d_counter += 1
        d_sum += row["Item_Price"]
    else:
        continue

print("Avg Price of food:",round(f_sum/f_counter, 2))
print("Avg Price of drink:",round(d_sum/d_counter, 2))
```

Ans6. Therefore the coffee shop makes more by selling drinks than by selling food
